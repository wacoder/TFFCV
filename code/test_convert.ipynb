```python
image, label = inputs("../data/tfrecords/*.tfrecords", 1,1)
output = alexnet_v2(image)
init_ops = tf.group(tf.global_variables_initializer(), tf.local_variables_initializer())
```

```python
with tf.Session() as sess:
    sess.run(init_ops)
    my_image, my_label = sess.run([image, output])
    
print(my_label.shape)
plt.imshow(my_image[0])
```
