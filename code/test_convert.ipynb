```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt 
from utils import inputs
from alexnet import alexnet_v2

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
image, label = inputs("../data/tfrecords/*.tfrecords", 1,1)
output = alexnet_v2(image)
init_ops = tf.group(tf.global_variables_initializer(), tf.local_variables_initializer())
```

```python
with tf.Session() as sess:
    sess.run(init_ops)
    my_image, my_label = sess.run([image, label])
    
print(my_label)
plt.imshow(my_image[0])
```
